# Representación de las clases del juego Blackjack

```python
class Juego: 
    # ATRIBUTOS
    menu: str
    cantidad_jugadores: int
    apuesta_min: int
    apuesta_max: int
    tipo_dealer: bool # maquina o jugador
    # RELACIONES
    baraja: Baraja
    jugador_es: Jugador
    dealer: Dealer
    # MÉTODOS
    def iniciar(): pass
    def pausar(): pass
    def reanudar(): pass
    def quitar(): pass
    
class Baraja: 
    # ATRIBUTOS
    tamaño: int
    tipo: str
    # RELACIONES
    cartas: list # porque son muchas 
    # MÉTODOS
    def mezclar(): pass

class Carta: 
    # ATRIBUTOS
    valor: int
    palo: str
    esta_oculta: bool
    # RELACIONES
    # MÉTODOS
    def mostrar(): pass

class Jugador:
    # ATRIBUTOS
    nombre: str
    detalles: str
    turno: bool
    # RELACIONES
    mano: Mano
    ficha_s: Fichas
    # MÉTODOS
    def pedir_carta(): pass
    def pasar(): pass
    def mostrar_carta(): pass
    def apostar(): pass

class Dealer:
    # ATRIBUTOS
    # RELACIONES
    mano: Mano
    # MÉTODOS
    def pedir_carta(): pass
    def pasar(): pass
    
class Mano:
    # ATRIBUTOS
    puntaje: int
    tamaño: int
    # RELACIONES
    cartas: list
    # MÉTODOS
    def mostrar_carta(): pass
    
class Fichas:
    # ATRIBUTOS
    valor: int
    # RELACIONES
    # MÉTODOS
    def nada(): pass
```
